# 5.class,module,예외처리
## 5-1. class
- C언어에는 class가 존재하지 않는다. class를 사용하지 않아도 프로그래밍 가능. 그렇다면 class는 왜 필요한 것인가?
- 한 프로그램에서 함수를 활용하여 여러가지 값을 기억해야 할 필요가 있을 경우, 같은 함수를 여러개 작성해야 하는 경우가 발생 -> class 활용!

- class예시)



- 클래스 vs 객체 vs 인스턴스

클래스   : 설계도면

객체     : 피조물(object)

인스턴스 : 피조물 보다는 관계를 설명

cf) 클래스 안에 구현된 함수 = 메서드

- 생성자 : 객체가 생성될 때 자동으로 호출되는 메서드
- 상속 : class 클래스이름(상속할 클래스 이름)
- 메서드 오버라이딩 : 부모 클래스에 있는 메서드를 동일한 이름으로 다시 만드는 것
