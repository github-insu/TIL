주어진 좌표가 각 사분면과 축에 점이 몇개 인지 구하기
1]첫 줄에 점의 개수(n)
2]다음 n개 줄에 점의 좌표(x,y)
3]축에 점이 몇개 인지도 출력

```python
n = int(input())
Q1,Q2,Q3,Q4,AXIS = 0,0,0,0,0
for i in range(n):
    x,y = map(int, input().split())
    if x>0 and y>0 :
        Q1 += 1
    elif x>0 and y<0:
        Q4 += 1
    elif x<0 and y>0:
        Q2 += 1
    elif x<0 and y<0:
        Q3 += 1
    else :
        AXIS += 1

print(f'Q1: {Q1}')
print(f'Q2: {Q2}')
print(f'Q3: {Q3}')
print(f'Q4: {Q4}')
print(f'AXIS: {AXIS}')
```

팰린드롬인지 확인
팰린드롬 = 앞으로 읽을 때와 거꾸로 읽을 때 똑같은 단어
정방향 순서 단어 == 역방향 순서 단어    -> 펠린드롬이면 1 , 아니면 0
역방향으로 어떻게 만들 것인지가 핵심

```python
#10988번 - 역방향 순서의 단어를 만들어서 원래단어와 비교해야한다는 로직
word = input()
reverse_word = ''
for i in range(1,len(word)+1):
    reverse_word += word[-i]

if word == reverse_word:
    print(1)
else:
    print(0)
```

광고를 해야 할지 말아야 할지 결정하는 프로르램
첫째 줄 = 테스트 케이스 개수(N)
다음 줄 = r(광고를 하지 않았을 때 수익), e(광고했을때 수익),c(광고 비용)

```python
N = int(input())
for i in range(N):
    r,e,c = map(int,input().split())
    if e-c < 0 :
        print('doe')
    if r < (e-c) :
        print('advertise')
    elif r > (e-c) :
        print('do not advertise')
    else :
        print('does not matter')
```

A와 B의 최소공배수 구하기
최소공배수 = (A * B)/최대공약수

```python
T = int(input())
result = 0
for i in range(T):
    A,B = map(int, input().split())
    gob = A*B
    multi = 1
    count = 2
    while True:
        if (A % count == 0) and (B % count == 0):
            multi *= count
            A /= count
            B /= count
        else :
            if count > min(A, B) :
                break
            count += 1
    result = int(gob/multi)
    print(result)
```
