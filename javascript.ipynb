## 변수
- 데이터 종류 : 문자형(String), 숫자형(Number), 논리형(Boolean), 빈(Null) 데이터
 - cf) 문자형과 숫자형을 연산하면 문자형 데이터가 출력
- 선언(var box;) : 변수명으로 데이터를 저장할 수 있는 공간이 생김
- 정의(box = 100;) : 변수에 값이 저장
- 변수에 값이 등록되기 전의 기본값(Undefined) & 변수에 값이 없는 경우 (Null)

## 연산자
- 종류 : 산술, 문자, 대입, 증감, 비교, 논리, 삼항조건연산자
- 증감 연산자 : var A = ++B  ----> B의 값을 연산 후 A에 대입
               var A = B++  ----> A에 B를 대입 후 연산 실행
- 비교 연산자 : ===(데이터 타입까지)
  - ex) 10 == '10' : True   /   10 === '10' : False
- 논리 연산자 : &&(and) , ||(or), !(not)
- 우선 순위 : 괄호 > 단항연산자 > 산술연산자 > 비교연산자 > 논리연산자 > 대입연산자
- 삼항 조건 연산자 : 조건식?참일때코드:거짓일때코드
  - ex) var result = a>b?'a':'b";

# 객체
- 자바스크립트는 객체(속성+기능)기반의 언어
- 종류 : 내장객체, 브라우저객체모델(BOM), 문서객체모델(DOM)

1. 내장객체
- 생성 : 참조변수(인스턴스이름) = new 생성함수()
  - var tv = new Object()
  - 기본객체(Object), 날짜객체(Date), 수학(Math.연산자()), 배열객체(Array), 문자열객체(String)

2. 브라우저 객체
- window는 브라우저 객체의 최상위 객체
- 그 하위에는 document, screen, location, history, navigator가 있음
 - screen객체 : 사용자 모니터 정보(속성)를 제공하는 객체
 - location객체 : 사용자 브라우저와 관련된 속성과 메서드를 제공하는 객체
 - history객체 : 사용자가 방문한 사이트의 기록을 남기고 이전 방문 사이트와 다음 방문 사이트로 다시 돌아갈 수 있는 속성과 메서드를 제공
 - navigator객체 : 현재 방문자가 사용하는 브라우저 정보와 운영체제 정보를 제공하는 객체

# 함수
- 프로그램의 코드를 저장한 공간
- 호이스팅 : 함수 정의문보다 호출문이 먼저 나와도 함수 정의문을 끌어올려 함수를 호출
  - 일반 함수 정의 방식은 호이스팅 적용, 그러나 익명 함수 선언 참조 방식은 적용x
- 매개변수 없이 함수에 전달된 값 받아오기
  - 함수 정의문의 매개변수가 없는 상태에서 데이터를 전달하여 함수를 호출하면 그 값은 배열에 저장(arguments라는 변수로 참조)
  - 정의문에 arguments; 작성
- 스코프
  - 의미 : 변수 또는 함수의 유효 범위
  - 종류 : 지역, 전역
  - 지역 변수 또는 지역 함수는 함수 스코프 내에서만 적용 가능
